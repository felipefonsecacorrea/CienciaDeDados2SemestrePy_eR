```python
# função cosseno

import math

#seno de 60 graus
print(math.cos(math.radians(60)))

print(math.cos(math.pi/3))
```

```python
import numpy as np

print(np.cos(np.pi/3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

fcos=np.cos(np.linspace(0,360,360) * np.pi/180)
x=np.linspace(0,360,360)

plt.figure().set_figheight(2,3)
plt.plot(x,fcos)
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.title('Função Cosseno',fontsize=14)
plt.legend(['f(x)=cos(x)'])
plt.show()
```

```python
# implemente uma função para obter a função cosseno abaixo e plote o recultado com um periodo ou mais
#f(x) = cos(2x) - 3
import numpy as np
import matplotlib.pyplot as plt

def fcos(x):
    return np.cos(2*x) - 3

plt.plot(fcos(np.linspace(0,2*np.pi,100)))
plt.xlabel('x',fontsize=13)
plt.ylabel('y',fontsize=13)
plt.title('Função Cosseno 2',fontsize=16)
plt.legend(['f(x) = cos(2x) - 1'])
plt.show()
```

```python
# função cosseno

import math

#seno de 60 graus
print(math.tan(math.radians(60)))

print(math.tan(math.pi/3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

ftan=np.tan(np.linspace(0,360,360) * np.pi/180)
x=np.linspace(0,360,360)

plt.figure().set_figheight(2,3)
plt.plot(x,ftan)
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.title('Função Cosseno',fontsize=14)
plt.legend(['f(x)=tan(x)'])
plt.ylim(-20,20)
plt.show()
```

```python
#escreva um codigo para plotar a função polinominal abaixo

import matplotlib.pyplot as plt

def poli(x):
    return (x**3) + (3*x) + 5

plt.plot(poli(np.linspace(-10,10,100)))
plt.xlabel('x',fontsize=13)
plt.ylabel('y',fontsize=13)
plt.title('Função Polinomial',fontsize=16)
plt.legend(['f(x) = x³ + 3x + 5'])
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x=np.linspace(0.1,10,30)
y=2.**x

plt.figure().set_figheight(2,3)
plt.plot(x, y)
plt.xlabel('x', fontsize = 15)
plt.ylabel('tan(x)', fontsize = 15)
plt.title('Funcao tan')
plt.legend(['f(x)=tan(x)'])
plt.slow=()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = np.linspace(0.1,10,30)
y = np.log(x)

plt.xlabel('x', fontsize = 15)
plt.ylabel('tan(x)', fontsize = 15)
plt.title('Funcao log')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x,y)
```
